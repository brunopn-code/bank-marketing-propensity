## Initial Data Inspection Findings

The dataset contains 41,188 rows and 21 columns. It includes 20 input features and one target variable, `y`, which indicates whether the client subscribed to a term deposit.

There are no missing values detected by pandas. However, some categorical variables contain the value `unknown`, which represents unavailable or undisclosed information.

The target variable is imbalanced. Most clients did not subscribe to the term deposit, with 36,548 `no` responses and 4,640 `yes` responses. This means that only about 11.27% of clients subscribed.

This class imbalance is important for modeling because accuracy alone may be misleading. The modeling stage should also evaluate metrics such as precision, recall, F1-score, ROC-AUC, and confusion matrix.

```python
df_clean = df.copy()
```

```python
df_clean["target"] = df_clean["y"].map({
    "yes": 1,
    "no": 0
})

df_clean = df_clean.rename(columns={
    "emp.var.rate": "emp_var_rate",
    "cons.price.idx": "cons_price_idx",
    "cons.conf.idx": "cons_conf_idx",
    "nr.employed": "nr_employed"
})

df_clean.head()
```

```python
df_clean.to_csv("bank_marketing_clean.csv", index=False)
```
